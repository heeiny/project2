```python
# 개별,연결 결측치 끼리 채우기 
for i in range(len(df)):
    if df.loc[i,'자산(*)(IFRS)(천원)'] == "NaN":
        df.loc[i,'자산(*)(IFRS)(천원)'] = df.loc[i,'자산(*)(IFRS연결)(천원)']
    else:
        if df.loc[i,'자본(*)(IFRS)(천원)'] == "NaN":
            df.loc[i,'자본(*)(IFRS)(천원)'] = df.loc[i,'자본(*)(IFRS연결)(천원)']
    # df.fillna(df.iloc[i,'자본율(개별)']) = df.iloc([i,'자본율(연결)'])
df.loc[14955:]
```

```python
# 개별,연결 결측치 끼리 채우기 
for i in range(len(df)):
    if df.iloc[i,'자산(*)(IFRS연결)(천원)'] ==" ":
        df.iloc[i,'자산(*)(IFRS연결)(천원)'] = df.iloc[i,'자산(*)(IFRS)(천원)']
    else:
        if df.iloc[i,'자본(*)(IFRS연결)(천원)'] == " ":
            df.iloc[i,'자본(*)(IFRS연결)(천원)'] = df.iloc[i,'자본(*)(IFRS)(천원)']
    # df.fillna(df.iloc[i,'자본율(개별)']) = df.iloc([i,'자본율(연결)'])
```

```python
df.to_csv("총자산대비자본.csv",encoding="CP949")
```
