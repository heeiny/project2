## 개별.연결 결측치 채우기

```python
# 개별 연결 합치기(매출액, 매출원가)
for i in range(len(df)):
    if df.loc[i,'자산(*)(IFRS)(천원)']=="NaN":
        df.loc[i,'자산(*)(IFRS)(천원)']=df.loc[i,'자산(*)(IFRS연결)(천원)']

for i in range(len(df)):
    if df.loc[i,'비유동금융부채(*)(IFRS)(천원)']=="NaN":
        df.loc[i,'비유동금융부채(*)(IFRS)(천원)']=df.loc[i,'비유동금융부채(*)(IFRS연결)(천원)']
df
```

### 총자산대비금융부채율 산식

```python
df1= df.drop(['자산(*)(IFRS)(천원)'],axis=1)
df1= df.drop(['비유동금융부채(*)(IFRS)(천원)'],axis=1)
df1.rename(columns={'자산(*)(IFRS연결)(천원)':'총자산'},inplace=True)
df1.rename(columns={'비유동금융부채(*)(IFRS연결)(천원)':'비유동금융부채'},inplace=True)
df1['총자산대비비유동금융부채율']=(df1["비유동금융부채"]/df1["총자산"])
df1
```

```python
# for i in range(len(df)-2):
#     df.loc[i,'총자산대비비유동금융부채율_'] = ((df.loc[i+1,'총자산대비비유동금융부채율_'] - df.loc[i+2,'총자산대비비유동금융부채율']) / df.loc[i+1,'총자산대비비유동금융부채율']) *100
# df
```

```python
df1['총자산대비비유동금융부채율_']=df1['총자산대비비유동금융부채율'].shift().where(df1.회사명.eq(df1.회사명.shift()))
df1['총자산대비유동금융부채율의변화율']=(df1['총자산대비비유동금융부채율']-df1['총자산대비비유동금융부채율_']/df1['총자산대비비유동금융부채율'])*100
df1['회사명'] = df1['회사명'].str.replace("(주)","")
df1['회사명'] = df1['회사명'].str.replace("\(","")
df1['회사명'] = df1['회사명'].str.replace("\)","")
# df1['회계년도'] = pd.to_datetime(df1['회계년도'],format='%b-%d').dt.strftime('20%d-%b')
df1
```

### 총자산대비금융부채율의 변화율 구하는 산식

```python
df1['총자산대비유동금융부채율의변화율_']=df1['총자산대비유동금융부채율의변화율'].shift().where(df1.회사명.eq(df1.회사명.shift()))
df1
```

```python
df1= df1.drop(['총자산','비유동금융부채'],axis=1)
```

```python
df1["변화율여부"]=""
```

```python
for i in range(0,len(df1)):
        if ((df1['총자산대비유동금융부채율의변화율'][i] >5 ) | (df1['총자산대비유동금융부채율의변화율_'][i] >5)):
                df1['변화율여부'][i] = 1
        else:
                df1['변화율여부'][i] = 0
```
