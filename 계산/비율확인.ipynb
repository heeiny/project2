### 전체 데이터프레임에서 행 개수가 2개 이상인 기업 추출하는 list생성 -> df_two

```python
# df_two는 행 개수가 2개 이상인 기업들을 추출하기 위한 list
# companylist는 전체 데이터 프레임에서 회사들의 갯수를 구하기 위한 list를 추출해야 함 
df_two = []
# company_list는 전체 데이터 프레임에서 개수가 2개 이상인 것들을 추출한 것
for i in company_list:
    if len(df[df['회사명']==i]) >= 3:
        df_two.append(i)
# 전체 데이터프레임에서 len(df_two)갯수
print(len(df_two))
print(df_two)
```

```python
# 행 개수가 2개 이상인 기업 인덱스
df_two_index = []
for i in df_two:
    df_two_index.append(df[df['회사명']==i].index)
df_two_index
```

```python
# 2년 연속 이자보상배율이 1미만인 기업을 발생 전년도부터 부실기업으로 정의
icr_count = 0
df['부실기업1'] = ''

for i in df_two_index:
    for j in i:
        if df.loc[j, '이자보상배율'] < 1:
            icr_count += 1
        else:
            icr_count = 0
            
        if icr_count >= 2:
            df.loc[j, '부실기업1'] = 1
        else:
            df.loc[j, '부실기업1'] = 0

for i in df.index:
    if df.loc[i, '부실기업1'] == '':
        df.loc[i, '부실기업1'] = 0
```
