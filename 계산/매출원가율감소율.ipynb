```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/user/Desktop/프로젝트2/데이터/data/회생전략/재무적 범주의 전략/비유동금융부채.csv',encoding='cp949')
df
```

```python
df['2012년 감소율'] = (((df['2011'] - df['2012'])/df['2011'])*100).round(2)
df['2013년 감소율'] = (((df['2012'] - df['2013'])/df['2012'])*100).round(2)
df['2014년 감소율'] = (((df['2013'] - df['2014'])/df['2013'])*100).round(2)
df['2015년 감소율'] = (((df['2014'] - df['2015'])/df['2014'])*100).round(2)
df['2016년 감소율'] = (((df['2015'] - df['2016'])/df['2015'])*100).round(2)
df['2017년 감소율'] = (((df['2016'] - df['2017'])/df['2016'])*100).round(2)
df['2018년 감소율'] = (((df['2017'] - df['2018'])/df['2017'])*100).round(2)
df['2019년 감소율'] = (((df['2018'] - df['2019'])/df['2018'])*100).round(2)
df
```

```python
df.fillna(0)
```

```python
# -20%이상 감소하면 0,1 입력하기

for i in range(0, len(df)):
    for year in range(2011, 2018):
        if year != 2019:
            if df.iloc[i][str(year+1)+"년 감소율"] < -20:
                df.loc[i, str(year+1)+"년 감소율여부"] = 1
            else:
                    df.loc[i, str(year+1)+"년 감소율여부"] = 0

        # else:
        #     if df.iloc[i][str(year)+"년 감소율"] < -20:
        #         df.loc[i, str(year)+"년 감소율여부"] = 1
        #     else:
        #             df.loc[i, str(year)+"년 감소율여부"] = 0
```

```python
df.to_csv("비금융부채감소여부.csv",encoding="CP949")
```
