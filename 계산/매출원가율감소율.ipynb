## 개별.연결 결측치 채우기

```python
# 개별 연결 합치기(매출액, 매출원가)
for i in range(len(df)):
    if df.loc[i,'자산(*)(IFRS)(천원)']=="NaN":
        df.loc[i,'자산(*)(IFRS)(천원)']=df.loc[i,'자산(*)(IFRS연결)(천원)']

for i in range(len(df)):
    if df.loc[i,'비유동금융부채(*)(IFRS)(천원)']=="NaN":
        df.loc[i,'비유동금융부채(*)(IFRS)(천원)']=df.loc[i,'비유동금융부채(*)(IFRS연결)(천원)']
df
```

## 개별 column삭제 후  d1프레임 생성/ 매출원가율 column생성

```python
df1 = df.drop(['매출액(수익)(*)(IFRS)(천원)'],axis=1,inplace=True)
df1 = df.drop(['매출원가(*)(IFRS)(천원)'],axis=1)
```

```python
df1.to_csv("C:/Users/user/Desktop/프로젝트2/데이터/data/회생전략/운용적.자산적 범주의 전략/매출원가축소/데이터중복프레임제거.csv",encoding='CP949')
```

```python
df1['매출원가율']=(df1['매출원가(*)(IFRS연결)(천원)']/df1['매출액(수익)(*)(IFRS연결)(천원)']).round(2)
df1
```

# 회사명column으로 매출원가 감소율 구하기

```python
#.shift()함수 사용 
#.where()함수 사용
```

```python
df1['매출원가율_'] = df1['매출원가율'].shift().where(df1.회사명.eq(df1.회사명.shift()))
df1['매출원가감소율']=(df1['매출원가율_']-df1['매출원가율'])
df1
```
