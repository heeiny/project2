```python
import pandas as pd
import numpy as np
```

```python
df_a= pd.read_csv(r"../data\dataset3\전체데이터_1년연속부실징후기업.csv",encoding='cp949')
df_a['회사명'] = df_a['회사명'].str.replace("(주)","")
df_a['회사명'] = df_a['회사명'].str.replace("\(","")
df_a['회사명'] = df_a['회사명'].str.replace("\)","")
df_a=df_a.sort_values(["회사명","회계년도"])
df_a
```

```python
df_b= pd.read_csv(r"../data\dataset3\매출원가감소율여부.csv",encoding='cp949')
df_b['회사명'] = df_b['회사명'].str.replace("(주)","")
df_b['회사명'] = df_b['회사명'].str.replace("\(","")
df_b['회사명'] = df_b['회사명'].str.replace("\)","")
df_b.rename(columns={"거래소코드_x":"거래소코드"},inplace=True)
df_b.drop('Unnamed: 0',axis=1,inplace=True)
df_b
```

## 비유동금융부채감소율여부

```python
df_f = pd.read_csv(r"../data\dataset3\비유동금융부채감소율여부.csv",encoding='cp949')
df_f.drop('Unnamed: 0',axis=1,inplace=True)
df_f=df_f.sort_values(["회사명","회계년도"])
df_f['회계년도']=pd.to_datetime(df_f['회계년도'],format='%b-%d').dt.strftime('20%d-%m')
df_f
```

```python
df_f = df_f.sort_values(['회계년도'])
df_f
```

```python
## 2010년에 들어가는 함수를 자를 수 있는 함수
t5 = df_f[df_f['회계년도'].str.contains('2010')].index
df_f = df_f.drop(t5)
df_f
```

```python
df_f = df_f.sort_values(['회사명','회계년도'])
df_f = df_f.sort_values(['회계년도'])
df_f
```

```python
df_f.to_csv("비유동여부.csv",encoding='cp949')
```
