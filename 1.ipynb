```python
import pandas as pd
import numpy as np
```

```python
df_a = pd.read_csv("data_fianl.csv", encoding="cp949")
df_b = pd.read_csv("매출원가율.csv", encoding='cp949')
df_a.drop(columns=["Unnamed: 0"], inplace=True)
df_b.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df_a=df_a.sort_values(["회사명","회계년도"])
df_a
```

```python
df_b = df_b.sort_values(["회사명","회계년도"])
df_b
```

```python
# merge("A","B",how="left")했는데 안돼서 A.merge(B,how="left")했더니 됨!^^ 
df3 = pd.merge(df_b, df_a, on=['회사명','회계년도'], how="left")
df3
```

```python
df3.rename(columns = {'채울열':'산업중간값'},inplace=True)
df3
```

```python
df3.to_csv("데이터프레임.csv",encoding='cp949')
```

```python
df_b = pd.read_csv("./데이터프레임.csv", encoding='cp949')
df_b.drop(columns=["Unnamed: 0"], inplace=True)
df_b
```

```python
df_b["매출원가감소율여부"]=""
df_b
```

```python
df_b=df_b.rename(columns={'산업중간값 여부':'산업중간값여부','감소율 여부':'감소율여부'})
df_b=df_b.fillna(0)
```

df.loc[행 인덱싱값,열 인덱싱 값]

```python
df_b['산업중간값여부']=df_b['산업중간값여부'].replace('#DIV/0!',0)
df_b['감소율여부']=df_b['감소율여부'].replace('#DIV/0!',0)
df_b['산업중간값여부']=df_b['산업중간값여부'].replace('#NAME?',0)
df_b['감소율여부']=df_b['감소율여부'].replace('#NAME?',0)
```

```python
df_b['산업중간값여부'].astype('int')
df_b['감소율여부'].astype('int')
```

```python
for i in range(0,len(df_b)):
        if (df_b.loc[i,'산업중간값여부'] == 1) | (df_b.loc[i,'감소율여부'] == 1):
                df_b.loc[i,'매출원가감소율여부'] = 1
        else:
                df_b.loc[i,'매출원가감소율여부'] = 0
```

```python
for i in range(0,len(df_b)):
        if ((df_b['산업중간값여부'][i] == 1) | (df_b['감소율여부'][i] == 1)):
                df_b['매출원가감소율여부'][i] = 1
        else:
                df_b['매출원가감소율여부'][i] = 0
```
